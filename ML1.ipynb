```python
# We are normalizing data
def normalize(data):
    result = data.copy()
    for i in data.columns:
        max_v = data[i].max()
        min_v = data[i].min()
        result[i] = (data[i] - min_v) / (max_v - min_v)
    return result
```

```python
# Назначаем все значение ка Х кроме СалеПрайс
X = train.drop(['SalePrice'], axis=1)
```

```python
# СайлПрайс присваиваем на у
y = train['SalePrice']
```

```python
# И находим длина параметра у
m = len(y)
```

```python
# Эта функция возвращает нормализованные значения, среднее значение, 
# и стандартное отклонение для каждой функции
def featureNormalize(X):
    X_norm = (X - np.mean(X))/np.std(X)
    mu = np.mean(X)
    sigma = np.std(X)
    
    return X_norm, mu, sigma
```

```python
#
X, mean, std = featureNormalize(X) 
```

```python
print(mean)
```

```python
X = np.append(np.ones([m,1]), X, axis=1)
```

```python
#
y = np.array(y).reshape(-1,1) ## reshape y to mx1 array.
```

```python
#
theta = np.zeros([76,1]) ## Initialize theta (the coefficient) to a 3x1 zero vector.
```

```python
#Эта функция принимает значения для обучающего набора, 
# а также начальные значения тета и возвращает стоимость 

def computeCostMulti(X,y, theta):
    m = len(y) 
    h = X.dot(theta)
    J = 1 / (2 * m) * (np.sum((h - y)**2))
    
    return J
```

```python
#
cost = computeCostMulti(X,y,theta) 

print(cost)
```

```python
#Эта функция принимает значения обучающего набора, а также исходные
# тета-значения (коэффициенты), скорость обучения и количество итераций. 
# Результатом будет новый набор коэффициентов (тета), оптимизированный для 
# прогнозирования, а также массив стоимости, которая обесценивается на каждой итерации. 
def gradientDescentMulti(X, y, theta, alpha, iter):
    J_history = [] 
    m = len(y) 
    for i in range(iter):
        h = X.dot(theta)
        theta = theta - (alpha / m) * (X.T.dot(h - y)) ## Gradient descent function
        J_history.append(computeCostMulti(X, y, theta)) 
    return theta, J_history
```

```python
# Вызвать функцию и передать параметры для вычисления нового коэффициента.
iter = 400 
alpha = 0.01

new_theta, J_history = gradientDescentMulti(X, y, theta, alpha, iter)

print (new_theta)
```

```python
# Мы снова вызываем функцию, но используем новые коэффициенты
new_cost = computeCostMulti(X,y,new_theta) 
```

```python
print(new_cost)
```

```python
#Эта функция учитывает характеристики дома, а также коэффициенты и 
#возвращает прогнозируемую цену
def prediction(X, theta):
    return np.dot(X, theta)
```

```python
pred = prediction(train,new_theta)
print(pred)
```
